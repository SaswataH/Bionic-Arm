## Importing req libraries



```python
import pandas as pd
```

## Dataset import

```python
df = pd.read_csv("https://raw.githubusercontent.com/SaswataH/Bionic-Arm/refs/heads/main/hand_data.csv")
df
```

## Checking the data shape

```python
# View the first few rows
print(df.head())

# Check the shape (rows , columns)
print("Shape: ",df.shape)

# Check column names and data types
print(df.info())
```

## Summary Statistics

```python
print(df.describe())
```

## Check for missing or duplicate data

```python
# Missing values
print(df.isnull().sum())

# Duplicates
print("Duplicate rows: ", df.duplicated().sum())
```

## Check Class Distribution

```python
print(df['label'].value_counts())
```

```python
import matplotlib.pyplot as plt

# Visualize the first hand sample in 2D
x_coords = [df.iloc[0][f'x{i}'] for i in range(21)]
y_coords = [df.iloc[0][f'y{i}'] for i in range(21)]

plt.figure(figsize=(5, 5))
plt.scatter(x_coords, y_coords, color='blue')
plt.gca().invert_yaxis()
plt.title(f"Label: {df.iloc[0]['label']}")
plt.xlabel("X")
plt.ylabel("Y")
plt.grid(True)
plt.show()
```
