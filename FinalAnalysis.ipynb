Finger Tips (4 , 8 , 12 , 16 , 20)

```python
# Landmarks for each finger tip
fingertip_landmarks = {
    'Thumb': '4',
    'Index': '8',
    'Middle': '12',
    'Ring': '16',
    'Pinky': '20'
}

# Axes and gesture list
axes = ['x', 'y', 'z']
gestures = df['label'].unique()
colors = ['#FF9999', '#99CCFF', '#99FF99']  # Customize per gesture

# Plot for each finger tip
for finger, lm_id in fingertip_landmarks.items():
    fig, ax = plt.subplots(figsize=(8, 5))

    # Collect grouped data for each axis
    all_data = []
    positions = []
    width = 0.2  # box width
    spacing = 1  # space between axis groups

    for i, axis in enumerate(axes):
        for j, gesture in enumerate(gestures):
            col = f'{axis}{lm_id}'
            gesture_data = df[df['label'] == gesture][col]
            all_data.append(gesture_data)
            # Position boxes: group at i * spacing, offset by gesture index
            positions.append(i * spacing + (j - 1) * width)

    # Plot manually positioned boxes
    bplot = ax.boxplot(all_data, positions=positions, widths=width, patch_artist=True)

    # Color each box based on gesture
    for i, box in enumerate(bplot['boxes']):
        box.set_facecolor(colors[i % len(gestures)])

    # Label x-ticks at center of each axis group
    ax.set_xticks([i * spacing for i in range(len(axes))])
    ax.set_xticklabels([axis.upper() for axis in axes])
    ax.set_title(f'{finger} Tip (Landmark {lm_id}) - Grouped Box Plot by Axis')
    ax.set_ylabel('Coordinate Value')
    ax.grid(axis='y', linestyle='--', alpha=0.4)

    # Add custom legend
    from matplotlib.patches import Patch
    legend_patches = [Patch(facecolor=colors[i], label=gestures[i]) for i in range(len(gestures))]
    ax.legend(handles=legend_patches, title='Gesture')

    plt.tight_layout()
    plt.show()
```

## Feature Extraction


```python
def euclidean_distance(p1, p2):
    return np.linalg.norm(np.array(p1) - np.array(p2))

def compute_angle(a, b, c):
    """Calculate angle ABC (in degrees) where B is the vertex"""
    ba = np.array(a) - np.array(b)
    bc = np.array(c) - np.array(b)
    cosine_angle = np.dot(ba, bc) / (np.linalg.norm(ba) * np.linalg.norm(bc))
    angle = np.arccos(np.clip(cosine_angle, -1.0, 1.0))
    return np.degrees(angle)

# Extract features
features = []
for _, row in df.iterrows():
    # Landmarks
    landmarks = np.array([[row[f'x{i}'], row[f'y{i}'], row[f'z{i}']] for i in range(21)])

    # Distance features
    dist_thumb_index = euclidean_distance(landmarks[4][:2], landmarks[8][:2])
    dist_wrist_index = euclidean_distance(landmarks[0][:2], landmarks[8][:2])
    dist_wrist_middle = euclidean_distance(landmarks[0][:2], landmarks[12][:2])

    # Spread of fingertips
    fingertip_indices = [4, 8, 12, 16, 20]
    fingertips_xy = landmarks[fingertip_indices, :2]
    var_x = np.var(fingertips_xy[:, 0])
    var_y = np.var(fingertips_xy[:, 1])

    # Mean and Std of coordinates
    mean_x = np.mean(landmarks[:, 0])
    std_x = np.std(landmarks[:, 0])
    mean_y = np.mean(landmarks[:, 1])
    std_y = np.std(landmarks[:, 1])
    mean_z = np.mean(landmarks[:, 2])
    std_z = np.std(landmarks[:, 2])

    # Angles
    angle_index_middle = compute_angle(landmarks[5][:2], landmarks[0][:2], landmarks[9][:2])

    features.append([
        dist_thumb_index, dist_wrist_index, dist_wrist_middle,
        var_x, var_y,
        mean_x, std_x, mean_y, std_y, mean_z, std_z,
        angle_index_middle,
        row['label']
    ])

# Create a new DataFrame with the extracted features
columns = [
    'dist_thumb_index', 'dist_wrist_index', 'dist_wrist_middle',
    'var_x_fingertips', 'var_y_fingertips',
    'mean_x', 'std_x', 'mean_y', 'std_y', 'mean_z', 'std_z',
    'angle_index_middle',
    'label'
]
features_df = pd.DataFrame(features, columns=columns)

features_df.to_csv("FeatureExt.csv", index=False)

features_df.head()
```

##Training NN

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder, StandardScaler
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.utils import to_categorical

# Load your features
features_df = pd.read_csv("FeatureExt.csv")  # Replace with your actual path

# Split features and labels
X = features_df.drop("label", axis=1).values
y = features_df["label"].values

# Encode string labels to integers
label_encoder = LabelEncoder()
y_encoded = label_encoder.fit_transform(y)

# Save classes for later prediction decoding
np.save("label_classes.npy", label_encoder.classes_)

# Normalize the feature values
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Save the scaler for live inference
import joblib
joblib.dump(scaler, "gesture_scaler.save")

# One-hot encode target labels
y_onehot = to_categorical(y_encoded)

# Train-test split
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y_onehot, test_size=0.2, random_state=42)

# Define ANN model
model = Sequential([
    Dense(64, activation='relu', input_shape=(X.shape[1],)),  # 11 features
    Dense(32, activation='relu'),
    Dense(y_onehot.shape[1], activation='softmax')  # Output neurons = number of classes
])

# Compile model
model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])

# Train model
history = model.fit(X_train, y_train, epochs=50, batch_size=16, validation_split=0.1)

# Evaluate model
loss, accuracy = model.evaluate(X_test, y_test)
print(f"Test accuracy: {accuracy:.2f}")

# Save the trained model
model.save("gesture_ann_model.h5")
```

```python
import matplotlib.pyplot as plt

plt.plot(history.history['accuracy'], label="Train Accuracy")
plt.plot(history.history['val_accuracy'], label="Val Accuracy")
plt.xlabel("Epochs")
plt.ylabel("Accuracy")
plt.title("Model Training Performance")
plt.legend()
plt.grid(True)
plt.show()
```
